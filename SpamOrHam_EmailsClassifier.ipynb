```python
import pandas as pd
from imblearn.over_sampling import RandomOverSampler
from collections import Counter
import numpy as np
```

```python
df=pd.read_csv('spam.csv')
df['spam']=(df['Category']=='spam').astype(int)
df=df.drop(['Category'],axis='columns')
df
```

```python
from sklearn.feature_extraction.text import CountVectorizer
v=CountVectorizer()
x_train_count=v.fit_transform(df['Message'].values)
x_train_count=x_train_count.toarray()
```

```python
cols=[f'word{i}' for i in range(8709)]
df1=pd.DataFrame(x_train_count)
df1.columns=cols
df1
```

```python
frames=[df,df1]
data=pd.concat(frames,axis=1, join='inner')
data['Spam']=data['spam']
data=data.drop(['Message','spam'],axis='columns')
data
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(data[data.columns[:-1]],data.Spam)
```

```python
from sklearn.naive_bayes import MultinomialNB
MNB_model=MultinomialNB()
MNB_model.fit(x_train,y_train)
```

```python
MNB_model.score(x_test, y_test)
```

```python
MNB_model.predict(x_test)
```
